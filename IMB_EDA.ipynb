```python
#Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
#Load in data
df_2000_data = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
df_2000_data
```

```python
#Load in data
df_2001_data = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')

df_2001_data
```

```python
#Concatenate
df = pd.concat((df_2000_data, df_2001_data))
df.info()
```

```python
# Filter out movies with both budget and revenue equal to 0
movies_revenue = df[(df['budget'] > 0) | (df['revenue'] > 0)]
movies_revenue
```

```python
# 1. How many movies had at least some valid financial information?
valid_movies_count = len(movies_revenue)
valid_movies_count
```

```python
# 2. How many movies are there in each certification category?
certification_counts = movies_revenue['certification'].value_counts()
certification_counts
```

```python
# 3. What is the average revenue per certification category?
average_revenue_by_certification = movies_revenue.groupby('certification')['revenue'].mean()
average_revenue_by_certification
```

```python
# 4. What is the average budget per certification category?
average_budget_by_certification = movies_revenue.groupby('certification')['budget'].mean()
average_budget_by_certification
```
