# Hypothesis 1

- Does the MPAA rating of a movie (G/PG/PG-13/R) affect how much revenue the movie generates?

- Null Hypothesis (H0): The MPAA rating of a movie does not have a statistically significant effect on how much revenue the movie generates.

- Alternative Hypothesis (H1): The MPAA rating of a movie has a statistically significant effect on how much revenue the movie generates.

```python
## Create groups dictionary. 
groups = {}
## Loop through all unique categories
for i in df['certification'].unique():
    ## Get series for group and rename
    data = df.loc[df['certification']==i,'revenue'].copy()
    
    # save into the dictionary
    groups[i] = data
groups.keys()
```

```python
norm_results = {}

for i, data in groups.items():
    if len(data) >= 8:  # Check if the sample size is at least 8
        stat, p = stats.normaltest(data)
        # Save the p val, test statistic, and the size of the group
        norm_results[i] = {'n': len(data),
                            'p': p,
                            'test stat': stat}
    else:
        # If sample size is less than 8, set values to NaN or skip the test
        norm_results[i] = {'n': len(data),
                            'p': None,
                            'test stat': None}

# Convert to a dataframe
norm_results_df = pd.DataFrame(norm_results).T

# Checking significance with pandas
norm_results_df['sig'] = norm_results_df['p'] < 0.05

# Display the results
print(norm_results_df)
```

- While our data doesn't follow a normal distribution, we can proceed with our analysis because each group contains more than 15 samples.

```python
# Extract data for G-rated and PG-rated movies
g_data = groups['G'].dropna()  # Drop NaN values if any
pg_data = groups['PG'].dropna()

# Perform a two-sample t-test
t_stat, p_value = stats.ttest_ind(g_data, pg_data, equal_var=False)  # Assuming unequal variances

# Define significance level (alpha)
alpha = 0.05

# Check if the p-value is less than alpha
if p_value < alpha:
    print("Reject the null hypothesis (H0): There is a significant difference in average movie revenue between G-rated and PG-rated movies.")
else:
    print("Fail to reject the null hypothesis (H0): There is no significant difference in average movie revenue between G-rated and PG-rated movies.")
```

```python
import matplotlib.pyplot as plt

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot([g_data, pg_data], labels=['G-rated', 'PG-rated'])
plt.title('Box Plot of Movie Revenue by Rating')
plt.xlabel('Rating')
plt.ylabel('Revenue')
plt.grid(True)

# Display the plot
plt.show()
```

# Hyphotesis 2

- Do movies that are over 2.5 hours long earn more revenue than movies that are 1.5 hours long (or less)?
- Null Hypothesis (H0): The length of a movie (over 2.5 hours) has no significant effect on its revenue.
- Alternative Hypothesis (H1): TMovies that are over 2.5 hours long earn significantly more revenue, on average, than movies that are 1.5 hours long or less.

```python
# Introducing a classification feature for movie length
df['is_long_movie'] = df['runtime'] > 150  # Indicates whether a movie is long (over 2.5 hours)
df['is_short_movie'] = df['runtime'] <= 90  # Indicates whether a movie is short (1.5 hours or less)
df.head()
```

```python
# Create separate DataFrames for long and short movies based on runtime
long_movies = df[df['runtime'] > 150][['revenue', 'runtime']]
short_movies = df[df['runtime'] <= 90][['revenue', 'runtime']]

# Display the counts of long and short movies
num_long_movies = len(long_movies)
num_short_movies = len(short_movies)

print(f"There are {num_long_movies} movies with a runtime of over 2.5 hours.")
print(f"There are {num_short_movies} movies with a runtime of 1.5 hours or less.")
```

```python
# Create the final revenue groups for long and short movies
revenue_long_movies = long_movies['revenue']
revenue_short_movies = short_movies['revenue']

# Display the first few records of the revenue groups for reference
display(revenue_long_movies.head(), revenue_short_movies.head())
```

```python
import numpy as np

# Detect outliers in the revenue of long movies
z_scores_long = stats.zscore(revenue_long_movies)
outliers_long = np.abs(z_scores_long) > 3
num_outliers_long = np.sum(outliers_long)

# Display the number of outliers found in the revenue of long movies
print(f"There are {num_outliers_long} outliers in the revenue of long movies.")
```

```python
import numpy as np

# Detect outliers in the revenue of short movies
z_scores_short = stats.zscore(revenue_short_movies)
outliers_short = np.abs(z_scores_short) > 3
num_outliers_short = np.sum(outliers_short)

# Display the number of outliers found in the revenue of short movies
print(f"There are {num_outliers_short} outliers in the revenue of short movies.")
```

```python
# Remove outliers from the revenue of short movies
revenue_short_movies = revenue_short_movies[(np.abs(stats.zscore(revenue_short_movies)) < 3)]
```

```python
# Perform a normality test on the revenue of the long movies
normality_test_result_long = stats.normaltest(revenue_long_movies)

# Display the results of the normality test
normality_test_result_long
```

```python
# Perform a normality test on the revenue of the short group
normality_test_result_short = stats.normaltest(revenue_short_movies)

# Display the results of the normality test
normality_test_result_short
```

 - Although the data is not normally distributed, we can proceed with the test because the sample sizes are sufficiently large.

```python
# Test for equal variance between the long and short groups' revenue
variance_test_result = stats.levene(revenue_long_movies, revenue_short_movies)

# Display the results of the variance test
variance_test_result
```

```python
# Perform an independent t-test assuming unequal variances between the long and short groups' revenue
t_test_result = stats.ttest_ind(revenue_long_movies, revenue_short_movies, equal_var=False)

# Display the results of the independent t-test
t_test_result
```

- The p-value obtained from the independent t-test is significantly less than the conventional significance level of 0.05. This suggests that we can confidently reject the null hypothesis. Therefore, we have strong evidence to conclude that there is a statistically significant difference in the revenue generated by short movies compared to long movies.

# Hyphotesis 3
- Do some genres earn more revenue than others?
- Null Hypothesis (H0): There is no significant difference in revenue between different movie genres. In other words, the average revenue of all genres is the same.
- Alternative Hypothesis (H1): At least one genre earns significantly more revenue than the others. In other words, there is a difference in the average revenue among different genres.

```python
# Create a contingency table to analyze the relationship between genres and revenue
genre_revenue_table = pd.crosstab(df['genres'], df['revenue'])
genre_revenue_table

```

```python
# Perform a chi-squared test and store the results
chi2_statistic, p_value, degrees_of_freedom, expected_values = stats.chi2_contingency(genre_revenue_table)

```

```python
# Determine the p-value obtained from the chi-squared test
print("The p-value from the chi-squared test is:", p_value)
```
